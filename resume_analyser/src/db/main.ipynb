```python
def find_smallest_integer(num_digits, sum_digits):
    if num_digits * 9 < sum_digits:
        return -1

    digits = [0] * num_digits
    remaining_sum = sum_digits

    for i in range(num_digits - 1, 0, -1):
        possible_digit = min(remaining_sum, 9)
        digits[i] = possible_digit
        remaining_sum -= possible_digit

    digits[0] = remaining_sum + 1

    return int(''.join(map(str, digits)))

```

```python
def digitSumZero(numDigits, sumDigits):
  if numDigits < 1 or numDigits > 18 or sumDigits < 1 or sumDigits > 200:
    return -1
  smallestInteger = 10 ** (numDigits - 1)
  for i in range(smallestInteger, 10 ** numDigits):
    if sumDigits == digitSum(i) and productDigits(i) == 0:
      return i
  return -1

def digitSum(n):
  sum = 0
  while n > 0:
    sum += n % 10
    n //= 10

  return sum

def productDigits(n):
  product = 1
  while n > 0:
    product *= n % 10
    n //= 10

  return product
```
